###6. Handling Null Values

```python
data = data.drop("Loan_ID",axis = 1)
```

```python
coloumn = ["Gender","Married","Dependents","LoanAmount","Loan_Amount_Term"]
```

```python
data = data.dropna(subset =coloumn)
```

```python
data["Self_Employed"].unique()
```

```python
data["Self_Employed"].mode()
```

```python
data["Self_Employed"]= data["Self_Employed"].fillna('No')
```

```python
data.isnull().sum()/len(data)*100
```

```python
data["Credit_History"].unique()
```

```python
data["Credit_History"].mode()
```

```python
data["Credit_History"]= data["Credit_History"].fillna(1.0)
```

###7. Handling Categorical Columns

```python
data.sample(6)
```

```python
data["Dependents"].unique()
```

```python
data["Dependents"]= data["Dependents"].replace(to_replace= '3+', value= '4')
```

```python
data["Dependents"].unique()
```

```python
data["Gender"].unique()
```

```python
data["Gender"] = data["Gender"].map({"Male" : 1 , "Female" : 0}).astype(int)
```

```python
data["Married"] = data["Married"].map({"Yes" : 1 , "No" : 0}).astype(int)
data["Education"] = data["Education"].map({"Graduate" : 1 , "Not Graduate" : 0}).astype(int)
data["Self_Employed"] = data["Self_Employed"].map({"Yes" : 1 , "No" : 0}).astype(int)
data["Property_Area"] = data["Property_Area"].map({"Rural" : 1 , "Urban" : 0 , "Semiurban" : 3}).astype(int)
data["Loan_Status"] = data["Loan_Status"].map({"Y" : 1 , "N" : 0}).astype(int)
```

```python
data.sample(5)
```

###8. Store Feature Matrix in X and Response(Target) in y

```python
X = data.drop('Loan_Status', axis = 1)
y = data['Loan_Status']
y
```

###9. Feature **Scaling**

```python
data.head()
```

```python
cols = ['ApplicantIncome', 'CoapplicantIncome' , 'LoanAmount', 'Loan_Amount_Term' ]
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X[cols] = ss.fit_transform(X[cols])
```

```python
X.head()
```

###10. Splitting Dataset into Training Set and Testing Set

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
def model_val(model,X,y):
    X_train,X_test,y_train,y_test=train_test_split(X,y,
                                                   test_size=0.20,
                                                   random_state=42)
    model.fit(X_train,y_train)
    y_pred=model.predict(X_test)
    print(f"{model} accuracy is {accuracy_score(y_test,y_pred)}")
```

###11. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model_val(model,X,y)
```

###12. SupportVectorClassifier (SVC)

```python
from sklearn.svm import SVC
model = SVC()
model_val(model,X,y)
```

###13. DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model_val(model,X,y)
```
